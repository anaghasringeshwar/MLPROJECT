```python
import pandas as pd

# Load the CSV with error handling
file_path = "styles.csv"  # Change this if your file is in a different location

try:
    df = pd.read_csv(file_path, error_bad_lines=False, warn_bad_lines=True)  # Ignore problematic lines
    print("File loaded successfully!")
except Exception as e:
    print("Error loading file:", e)

# Display first few rows
df.head()
```

```python
# Check basic information about the dataset
df.info()

# Show first few rows
df.head()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Drop rows where essential columns have missing values
df_cleaned = df.dropna(subset=["baseColour", "season", "productDisplayName"])

# Fill missing 'year' with the most common year (mode)
df_cleaned["year"].fillna(df_cleaned["year"].mode()[0], inplace=True)

# Fill missing 'usage' with "Unknown"
df_cleaned["usage"].fillna("Unknown", inplace=True)

# Check if missing values are handled
df_cleaned.isnull().sum()
```

```python
# Re-load the dataset to avoid modifications on a slice
df = pd.read_csv("styles.csv")

# Drop rows where essential columns have missing values
df_cleaned = df.dropna(subset=["baseColour", "season", "productDisplayName"]).copy()

# Fill missing 'year' safely
df_cleaned.loc[:, "year"] = df_cleaned["year"].fillna(df_cleaned["year"].mode()[0])

# Fill missing 'usage' safely
df_cleaned.loc[:, "usage"] = df_cleaned["usage"].fillna("Unknown")

# Check if missing values are handled
df_cleaned.isnull().sum()
```
